```python
#importing important libraries:

import requests
from bs4 import BeautifulSoup
```

```python
# request is used to pull the URL of the website:
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
req = requests.get(url)
```

```python
#start scraping:
soup = BeautifulSoup(req.content, 'lxml')
Product_title = soup.find('a', class_ = 'title').text
price = soup.find('h4', class_ = 'price float-end card-title pull-right').text
des = soup.find('p', class_ = 'description card-text').text
print(Product_title)
print(price)
print(des)
```

```python
#This block scaping all the information from the test website:
information = soup.find_all('div', class_ = 'col-md-4 col-xl-4 col-lg-4')
for products in information:
    titles = products.a.text
    prices = products.h4.text
    descs = products.p.text
    print(titles)
    print(prices)
    print(descs)
    print()
```
